```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Load dataset
file_path = "./data/SOCR-HeightWeight.csv"
df = pd.read_csv(file_path)

# Extract relevant columns
X = df[["Height(Inches)"]].values  # Feature (height)
y = df["Weight(Pounds)"].values  # Target (weight)

# Fit linear regression model
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)

# Calculate MSE loss for comparison
mse_loss = mean_squared_error(y, y_pred)
print(f"Linear Regression MSE Loss: {mse_loss:.4f}")

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X.flatten(), y=y, label="Actual Data", alpha=0.5)
plt.plot(X, y_pred, color='red', label="Linear Regression")
plt.xlabel("Height (Inches)")
plt.ylabel("Weight (Pounds)")
plt.title("Height vs Weight Scatter Plot with Regression Line")
plt.legend()
plt.show()
```
